## Вопрос 4

Какое число ближе к средней длительности фильма в датасете?

### Варианты ответов
- 115
- 110
- 105
- 120
- 100

```python
answers = [115, 110, 105, 120, 100]
```

```python
data_avg = data.runtime.mean()
result = answers[0]
```

```python
for a in answers:
    if abs(result - data_avg) > abs(a - data_avg):
        result = a
```

```python
result
```

```python
answers[[abs(x - data_avg) for x in answers].index(min([abs(x - data_avg) for x in answers]))]
```
